```python
import pandas as pd
from fuzzywuzzy import process

# Read the first table
table1 = pd.read_csv('/content/BPR.csv',encoding='latin1')

# Read the second table
table2 = pd.read_csv('/content/BPR_CK.csv',encoding='latin1')
# Function to find the best match for a topic in table1
def find_best_match(topic, choices):
    return process.extractOne(topic, choices)

# Iterate through each row in the second table and find the best match in the first table
table2['Lecture Code'] = table2['ï»¿Lecture Content'].apply(lambda x: find_best_match(x, table1['Lecture Content']))

# Extract Lecture Code of HO from the best match
table2['Lecture Code'] = table2['Lecture Code'].apply(lambda x: x[2] if x else None)

# Drop unnecessary columns
table2 = table2.drop(['Best_Match'], axis=1)

# Save the result to a new CSV file
table2.to_csv('content/merged_table.csv', index=False)
```
