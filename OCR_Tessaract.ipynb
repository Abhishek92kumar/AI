```python
import pandas as pd

# Our arguments
WIDTH_IMAGE = 800  # Width of the input image
HEIGHT_IMAGE = 504  # Height of the input image
TESSERACT_OUTPUT_FILEPATH = 'file1.csv'  # Path to the output TSV file created by Tesseract
# The ranges (min-max) of acceptable values for relative width/height/x/y values
MIN_VALUES = {
    'width': 0.14,
    'height': 0.03,
    'x': 0.45,
    'y': 0.50
}
MAX_VALUES = {
    'width': 0.18,
    'height': 0.05,
    'x': 0.60,
    'y': 0.65
}

# Define list for outputs
OUTPUTS = []

# Read the tab-separated ("\t") values from the file
df = pd.read_csv(TESSERACT_OUTPUT_FILEPATH, sep="\t")

# Iterate through words (level==5)
for _, word in df[df["level"]==5].iterrows():

    # Relative size of the word within the image
    relative_width = word["width"] / WIDTH_IMAGE
    relative_height = word["height"] / HEIGHT_IMAGE

    # Relative position of the word within the image
    relative_x = word["left"] / WIDTH_IMAGE
    relative_y = word["top"] / HEIGHT_IMAGE

    # If the relative values are not within set limits, continue to the next word
    if relative_width>MAX_VALUES['width']:
        continue
    if relative_width<MIN_VALUES['width']:
        continue
    if relative_height>MAX_VALUES['height']:
        continue
    if relative_height<MIN_VALUES['height']:
        continue
    if relative_x>MAX_VALUES['x']:
        continue
    if relative_x<MIN_VALUES['x']:
        continue
    if relative_y>MAX_VALUES['y']:
        continue
    if relative_y<MIN_VALUES['y']:
        continue

    # If all validation rules were passed, you have a potential match!
    OUTPUTS.append(word['text'])

# Print out all the matches!
for word in OUTPUTS:
    print(word)
```
